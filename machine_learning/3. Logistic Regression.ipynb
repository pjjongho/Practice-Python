##### 로지스틱 회귀(Logistic Regression)

- 로지스틱 회귀는 이름에 회귀라는 단어가 들어가지만, <span style = "color:pink">가능한 클래스가 2개인 이진 분류를 위한 모델</span>
- 로지스틱 회귀의 예측 함수 정의

<div align="center">

$\displaystyle \sigma(x) = \frac{1}{1+e^{-x}}$

$\displaystyle \hat{y} = \sigma(w_0 + w_1x_1 + . . . + w_px_p)$

</div>

- $\sigma$ : 시그모이드 함수
- 로지스틱 회귀 모델은 선형 회귀 모델에 시그모이드 함수를 적용
- 로지스틱 회귀의 학습 목표는 다음과 같은 **<span style="color:pink">목적 함수를 최소화 하는 파라미터 w를 찾는 것</span>**

<div align="center">

$\displaystyle BinaryCrossEntropy = - \frac{1}{N} \sum_{i=1}^N y_i \log (\hat{y_i}) + (1 - y_i) \log(1-\hat{y_i})$

</div>

##### 로지스틱 회귀 예제

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
samples = 1000
X, y = make_classification(n_samples=samples, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6))

ax.grid()
ax.set_xlabel('X')
ax.set_ylabel('y')

for i in range(samples):
    if y[i] ==0 :
        ax.scatter(X[i,0], X[i, 1], edgecolors='k', alpha=0.5, marker='^', color='r')
    else:
        ax.scatter(X[i, 0], X[i, 1], edgecolors='k', alpha=0.5, marker='v', color='b')

plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print(f'Scores of Training Data : {model.score(X_train, y_train)}')
print(f'Scores of Test Data : {model.score(X_test, y_test)}')
```

```python
scores = cross_val_score(model, X, y, scoring='accuracy', cv=10)
print(f'CV mean score : {scores.mean()}')
```

```python
model.intercept_, model.coef_
```

```python
x_min, x_max = X[:,0].min() -0.5, X[:,0].max() + 0.5
y_min, y_max = X[:,1].min() -0.5, X[:,1].max() + 0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(10,6))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Pastel1)

plt.scatter(X[:,0], X[:,1], c=np.abs(y-1), edgecolors='k', alpha=0.5, cmap=plt.cm.coolwarm)
plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks()
plt.yticks

plt.show()
```

##### 붓꽃 데이터에 대한 로지스틱 회귀

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data[:, [2,3]], iris.target, test_size=0.2, random_state=42, stratify=iris.target)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs',multi_class='auto', C=100.0, random_state=42)
model.fit(X_train, y_train)
```

```python
print(f'Scores of Training Data : {model.score(X_train, y_train)}')
print(f'Scores of Test Data : {model.score(X_test, y_test)}')
```

```python
X = np.vstack((X_train, X_test))
y = np.hstack((y_train, y_test))
```

```python
a = np.arange(10, 50, 1)
d = np.meshgrid(a)
print(f'a : {a}')
print(f'b : {d}')
```

```python
from matplotlib.colors import ListedColormap

x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1
x2_min, x2_max = X[:, 1].min()-1, X[:, 1].max()+1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02),
                       np.arange(x2_min, x2_max, 0.02))

Z = model.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
Z = Z.reshape(xx1.shape)

species = ('Setosa', 'Versicolour', ' Virginica')
markers = ('^', 'v','s')
colors = ('blue','purple','red')
cmap = ListedColormap(colors[:len(np.unique(y))])
plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
plt.xlim(xx1.min(), xx1.max())
plt.ylim(xx2.min(), xx2.max())

for idx, cl in enumerate(np.unique(y)):
    plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=colors[idx], marker=markers[idx], label=species[cl], edgecolors='k')

X_comb_test, y_comb_test = X[range(105, 150), :], y[range(105, 150)]
plt.scatter(X_comb_test[:, 0], X_comb_test[:, 1],
            c='yellow', edgecolors='k', alpha=0.2, linewidths=1, marker='o', s=100, label='Test')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(loc='upper left')
plt.tight_layout();
```

##### 확률적 경사 하강법 (Stochastic Gradient Descent)

- 모델을 학습 시키기 위한 간단한 방법
- 학습 파라미터에 대한 손실 함수의 기울기를 구해 기울기가 최소화 되는 방향으로 학습

<div align="center">

$\displaystyle \frac{\partial L}{\partial w} = \lim_{h \to 0} \frac{L(w+h) - L(w)}{h}$

$\displaystyle w'= w - \alpha \frac{\partial L}{\partial w}$

</div>

- scikit-learn에서는 <span style = "color:pink"> 선형 SGD 회귀와 SGD 분류를 지원함 </span>

##### SGD를 사용한 선형 회귀 분석

```python
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston_df = pd.read_csv(url)
print(boston_df.head())
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(boston_df.drop('medv',axis=1), boston_df['medv'], test_size=0.2)
```

```python
model = make_pipeline(StandardScaler(), SGDRegressor(loss='squared_error')) # 선형회귀할때는 'squared error'를 주로 사용함
model.fit(X_train, y_train)
```

```python
print(f'Scores of Training Data : {model.score(X_train, y_train)}')
print(f'Scores of Test Data : {model.score(X_test, y_test)}')
```

##### Iris 데이터에 대한 SGD 분류
